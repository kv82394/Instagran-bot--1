# 6 Extract list of followers

### 6.1 Extract the usernames of the first 500 followers of â€˜foodtalkindiaâ€™ and â€˜sodelhiâ€™

#### extracting for foodtalkindia

```python
#searching the text
text_box = driver.find_element_by_class_name('XTCLo') 
text_box.send_keys('foodtalkindia')
driver.implicitly_wait(10)
#clicking search
btn = driver.find_element_by_class_name('Ap253')
btn.click()
#clicking on followers btn
driver.implicitly_wait(10)
btn = driver.find_elements_by_class_name('-nal3 ')
btn[1].click()
driver.implicitly_wait(10)
#followers list
li = []
while True:
        driver.find_element_by_class_name("isgrP").click() ## switching to followers list 
        driver.find_element_by_tag_name("body").send_keys(Keys.END) ### scrolling using End Key
        li=driver.find_elements_by_class_name("wo9IH")
        if len(li)>=500:
            break
for i  in range(500):
    print((i+1),"->",(li[i].text.strip()).split("\n")[0].strip())
```
